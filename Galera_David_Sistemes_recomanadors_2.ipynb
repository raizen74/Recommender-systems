```python
import pandas as pd
import numpy as np
ratings = pd.read_csv("/content/drive/MyDrive/Advanced ML/ratings.csv")
print('ratings shape:',ratings.shape)
print(ratings.columns)
ratings.head()

```

```python
movies = pd.read_csv("/content/drive/MyDrive/Advanced ML/movies.csv")
print('movies shape:',movies.shape)
print(movies.columns)
movies.head()
```

Inserim una columna adicional amb el t√≠tol de la pel¬∑l√≠cula

```python
ratings = pd.merge(ratings,movies[['movieId','title']],on=['movieId'],how='left')
#print(ratings.dtypes)
ratings.head()
```

```python
print("N√∫mero d'usuaris:",len(ratings.userId.unique()))#610 usuaris
print("N√∫mero de pel¬∑l√≠cules:",len(ratings.title.unique()))#9719 pel¬∑l√≠cules
```

Creem un DataFrame on cada inst√†ncia representa un usuari, cada columna representa una pel¬∑l√≠cula i els valors corresponen als ratings dels l'usuaris a les pel¬∑l√≠cules. Les pel¬∑l√≠cules no puntuades pels usuaris les omplim amb un valor de -1.

```python
pivot=ratings.pivot_table(values='rating',index='userId',columns='title',fill_value=-1)
pivot.reset_index(drop=True,inplace=True)
pivot.head()
```

Podem veure per exemple com el primer usuari ha votat 232 pel¬∑l√≠cules amb un *rating* mitj√† de 4.36

```python
print(pivot.iloc[0,:][pivot.iloc[0,:]>=0].describe())
```

Calculem el rating promig de tot el DataFrame.

```python
ratings.rating[ratings.rating>=0].describe()
```

Establim un rating m√≠nim de 2.5 (equivalent a 5 sobre 10) per classificar la pel¬∑l√≠cula com *puntuada favorablement* per l'usuari. √âs a dir, si cada usuari representa una transacci√≥, la pel¬∑l√≠cula en q√ºesti√≥ forma part de la transacci√≥ si ha estat puntuada amb un rating>=2.5.

```python
llindar = 2.5
pivot = pivot.transform(lambda x: x >= llindar)
print(pivot.iloc[0,:].sum())
pivot.head()
```

Extraiem els itemsets frequents del DataFrame que apareixen com a m√≠nim en un 20% de les transaccions utilitzant l'algorisme *apriori*

```python
from mlxtend.frequent_patterns import apriori
freq_itemsets = apriori(pivot,min_support=0.2,use_colnames=True)
print(freq_itemsets.sort_values('support',ascending=False).tail(30))
print(freq_itemsets.shape)
```

Definim les regles d'associaci√≥ d'inter√©s quan la confian√ßa>=0.5

```python
from mlxtend.frequent_patterns import association_rules
asrules = association_rules(freq_itemsets,metric = "confidence", min_threshold = 0.5)
print(asrules.shape)
asrules.head(12)
```

# Recomanaci√≥ de pel¬∑l√≠cules donat el t√≠tol d'una o de v√†ries pel¬∑l√≠cules.

Defineixo una funci√≥ que retorna una llista amb els t√≠tols introdu√Øts per l'usuari dels quals vol recomanacions similars. En l'exemple, busquem pel¬∑l√≠cules similars a *American Beauty (1999)*

```python
def pelicules_similars():
  print('Pel¬∑l√≠cules similars a:')
  print("-------------------------")
  titols_similars_a = input("Introdueix una o v√†ries pel¬∑l√≠cules separades per coma:")
  #American Beauty (1999)
  my_list = titols_similars_a.split(",")
  if all(x in list(movies.title) for x in my_list) == False:
      print('T√≠tol/s erroni')
  else:
      return my_list

similar_titles_to = pelicules_similars()
```

Busquem dins el DataFrame de regles d'associaci√≥ les regles on tots els *antecedents* formen part de la llista **similar_titles_to** i recomanem els t√≠tols de la columna *consequents* sempre i quan aquests no figurin a la llista **similar_titles_to**.

```python
def troba_titols_per_recomanar(asrules, titles_list):
    titols = []
    for i in asrules.index:#itera entre totes les regles d'associaci√≥
        if asrules["antecedents"][i].issubset(frozenset(similar_titles_to)):#filtra les regles d'associaci√≥ on TOTS els antecedents s√≥n un subset de la llista similar_titles_to
            titols.extend(list(asrules["consequents"][i]))#afegeix els t√≠tols conseq√ºents a la regla     
    return set(titols).difference(set(similar_titles_to))#retorna els t√≠tols que s√≥n conseq√ºents i que no ha introdu√Øt l'usuari
titols_recomanats = troba_titols_per_recomanar(asrules, similar_titles_to)
print(len(titols_recomanats))
```

Imprimim els t√≠tols recomanats a l'usuari.

```python
def escriu_titols(llista_titols):
    for titol in llista_titols:
        print(titol)
        print("-------------------------")

escriu_titols(titols_recomanats)
```
