Inserim una columna adicional amb el títol de la pel·lícula

```python
ratings = pd.merge(ratings,movies[['movieId','title']],on=['movieId'],how='left')
#print(ratings.dtypes)
ratings.head()
```

```python
print("Número d'usuaris:",len(ratings.userId.unique()))#610 usuaris
print("Número de pel·lícules:",len(ratings.title.unique()))#9719 pel·lícules
```

Creem un DataFrame on cada instància representa un usuari, cada columna representa una pel·lícula i els valors corresponen als ratings dels l'usuaris a les pel·lícules. Les pel·lícules no puntuades pels usuaris les omplim amb un valor de -1.

```python
pivot=ratings.pivot_table(values='rating',index='userId',columns='title',fill_value=-1)
pivot.reset_index(drop=True,inplace=True)
pivot.head()
```

Podem veure per exemple com el primer usuari ha votat 232 pel·lícules amb un *rating* mitjà de 4.36

```python
print(pivot.iloc[0,:][pivot.iloc[0,:]>=0].describe())
```

Calculem el rating promig de tot el DataFrame.

```python
ratings.rating[ratings.rating>=0].describe()
```

Establim un rating mínim de 2.5 (equivalent a 5 sobre 10) per classificar la pel·lícula com *puntuada favorablement* per l'usuari. És a dir, si cada usuari representa una transacció, la pel·lícula en qüestió forma part de la transacció si ha estat puntuada amb un rating>=2.5.

```python
llindar = 2.5
pivot = pivot.transform(lambda x: x >= llindar)
print(pivot.iloc[0,:].sum())
pivot.head()
```

Extraiem els itemsets frequents del DataFrame que apareixen com a mínim en un 20% de les transaccions utilitzant l'algorisme *apriori*

```python
from mlxtend.frequent_patterns import apriori
freq_itemsets = apriori(pivot,min_support=0.2,use_colnames=True)
print(freq_itemsets.sort_values('support',ascending=False).tail(30))
print(freq_itemsets.shape)
```

Definim les regles d'associació d'interés quan la confiança>=0.5

```python
from mlxtend.frequent_patterns import association_rules
asrules = association_rules(freq_itemsets,metric = "confidence", min_threshold = 0.5)
print(asrules.shape)
asrules.head(12)
```

# Recomanació de pel·lícules donat el títol d'una o de vàries pel·lícules.

Defineixo una funció que retorna una llista amb els títols introduïts per l'usuari dels quals vol recomanacions similars. En l'exemple, busquem pel·lícules similars a *American Beauty (1999)*

```python
def pelicules_similars():
  print('Pel·lícules similars a:')
  print("-------------------------")
  titols_similars_a = input("Introdueix una o vàries pel·lícules separades per coma:")
  #American Beauty (1999)
  my_list = titols_similars_a.split(",")
  if all(x in list(movies.title) for x in my_list) == False:
      print('Títol/s erroni')
  else:
      return my_list

similar_titles_to = pelicules_similars()
```

Busquem dins el DataFrame de regles d'associació les regles on tots els *antecedents* formen part de la llista **similar_titles_to** i recomanem els títols de la columna *consequents* sempre i quan aquests no figurin a la llista **similar_titles_to**.

```python
def troba_titols_per_recomanar(asrules, titles_list):
    titols = []
    for i in asrules.index:#itera entre totes les regles d'associació
        if asrules["antecedents"][i].issubset(frozenset(similar_titles_to)):#filtra les regles d'associació on TOTS els antecedents són un subset de la llista similar_titles_to
            titols.extend(list(asrules["consequents"][i]))#afegeix els títols conseqüents a la regla     
    return set(titols).difference(set(similar_titles_to))#retorna els títols que són conseqüents i que no ha introduït l'usuari
titols_recomanats = troba_titols_per_recomanar(asrules, similar_titles_to)
print(len(titols_recomanats))
```
